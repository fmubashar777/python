```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x+1, color='blue', linewidth=0.25)
ax.plot(x,x+2, color='blue', linewidth=0.5)
ax.plot(x,x+3, color='blue', linewidth=1)
ax.plot(x,x+4, color='blue', linewidth=2)

ax.plot(x,x+5, color='green', linewidth=2,ls='-')
ax.plot(x,x+6, color='green', linewidth=3,ls='-.')
ax.plot(x,x+7, color='green', linewidth=4,ls=':')
ax.plot(x,x+8, color='green', linewidth=3,ls='-')


line,=ax.plot(x,x+9,color='black',lw=1.5)
line.set_dashes([5,10,15])

ax.plot(x ,x+10, color="orange",lw=2,ls='-',marker='+', markersize=8)
ax.plot(x ,x+11, color="pink",lw=2,ls='-',marker='.', markersize=18, markerfacecolor='red',)
```

```python
#Axis Apperance
```

```python
fig,axes=plt.subplots(1,3,figsize=(12,4))
axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default axes ranges')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylabel([0, 60])
axes[2].set_xlabel([2, 5])
axes[2].set_title('custom axes ranges')
```
